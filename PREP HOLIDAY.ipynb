```python
import pandas as pd
import numpy as np

# =========================
# STEP 1: Load holiday data
# =========================
holidays = pd.read_csv("holidays_events.csv")

print(holidays.head())
print(holidays.info())
```

```python
# =========================
# STEP 2: Normalize column names
# =========================

# Convert date to datetime
holidays["date"] = pd.to_datetime(holidays["date"])

# Convert transferred to bool
holidays["transferred"] = holidays["transferred"].astype(bool)
```

```python
# =========================
# STEP 3: Filter out transferred holidays
# =========================

holidays = holidays[holidays["transferred"] == False]
```

```python
# =========================
# STEP 4: Filter valid event types
# =========================

valid_types = ["Holiday", "Event"]
holidays = holidays[holidays["type"].isin(valid_types)]
```

```python
# =========================
# STEP 5: Keep national holidays
# =========================

holidays = holidays[holidays["locale"] == "National"]
```

```python
# =========================
# STEP 6: Create is_holiday variable
# =========================

holidays["is_holiday"] = 1
```

```python
# =========================
# STEP 7: Keep necessary columns
# =========================

holidays_clean = holidays[["date", "is_holiday"]]
```

```python
# =========================
# STEP 8: Group by date
# =========================

holidays_clean = (
    holidays_clean
    .groupby("date", as_index=False)
    .max()
)
```

```python
# =========================
# STEP 9: Save cleaned data
# =========================

holidays_clean.to_csv("holidays_cleaned.csv", index=False)
holidays_clean.head()
```
