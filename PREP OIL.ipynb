```python
import pandas as pd
import numpy as np

# =========================
# STEP 1: Load oil data
# =========================
oil = pd.read_csv("oil.csv")

print(oil.head())
print(oil.info())
```

```python
# =========================
# STEP 2: Normalize column names
# =========================

oil["date"] = pd.to_datetime(oil["date"])
oil = oil.rename(columns={"dcoilwtico": "oil_price"})
```

```python
# =========================
# STEP 3: Sort by date
# =========================

oil = oil.sort_values("date").reset_index(drop=True)
```

```python
# =========================
# STEP 4: Check for missing values
# =========================
print(oil.isna().sum())
```

```python
# =========================
# STEP 5: Handle missing oil_price
# =========================

# Forward fill the missing oil prices
oil["oil_price"] = oil["oil_price"].ffill()
```

```python
# =========================
# STEP 6: Check continuity
# =========================

date_diff = oil["date"].diff().dt.days.value_counts()
print(date_diff.head())
```

```python
# =========================
# STEP 7: Create lag columns for oil_price (optional)
# =========================
oil["oil_price_lag_1"] = oil["oil_price"].shift(1)
oil["oil_price_lag_7"] = oil["oil_price"].shift(7)
```

```python
# =========================
# STEP 8: Save cleaned data
# =========================
oil.to_csv("oil_cleaned.csv", index=False)
oil.info()
```
